```python
import altair as alt
import pandas as pd

alt.data_transformers.disable_max_rows()

df = pd.read_csv('https://github.com/UIUC-iSchool-DataViz/is445_data/raw/main/licenses_fall2022.csv')
df
```

```python
df.isna().sum()
```

```python
licenses = df[['License Type', 'License Status', 'County']].dropna()
licenses
```

```python
license_type = licenses['License Type'].value_counts().reset_index()
license_type.columns = ['License Type', 'Count']
license_type
```

```python
plot1 = alt.Chart(license_type).mark_bar().encode(
    x = alt.X('Count:Q', title = 'Num of Licenses'),
y = alt.Y('License Type:N', sort = '-x', title = 'License Type'),
color = alt.value('steelblue'),
tooltip = ['License Type', 'Count'],
).properties(width = 600, height = 400, title='Num of Licenses by License Types')
plot1
```

```python
plot1.save('license_types.json')
plot1.save('license_types.html')
```

```python
# I used ChatGPT 5.4 to help me with printing the county names correctly. Initially, no values were printing out because I didn't include .index so it was returning nothing.
county = licenses.dropna(subset = ['County'])
status = county.groupby(['County', 'License Status']).size().reset_index(name = 'Count')
top_10 = county['County'].value_counts().head(10).index
status = status[status['County'].isin(top_10)]
status
```

```python
dropdown =  alt.binding_select(
    options = list(status['License Status']), name = 'License Status: ')
select = alt.param( bind = dropdown, value = status['License Status'])

plot2 = alt.Chart(status).add_params(select).transform_filter(alt.datum['License Status'] == select).mark_bar().encode(
    x = alt.X('Count:Q', title = 'Num of Licenses'), y = alt.Y('County:N', sort = '-x', title = 'County'),
    color = alt.value('steelblue'),
    tooltip = ['County', 'License Status', 'County']).properties(
    width = 600, height = 400, title = 'Top Counties Listed by License Status')
plot2
```

```python
plot2.save('status.json')
plot2.save('status.html')
```
